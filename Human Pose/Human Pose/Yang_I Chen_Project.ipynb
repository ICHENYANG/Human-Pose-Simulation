```python
# Mean, Median, Max, Min, Stdev
neighbors= range(0,42240,480)
Mean=[]
Median=[]
Max=[]
Min=[]
Std=[]

for i, k in enumerate(neighbors):
    mean=X.iloc[0+k:480+k].mean()
    median=X.iloc[0+k:480+k].median()
    maximum=X.iloc[0+k:480+k].max()
    minimum=X.iloc[0+k:480+k].min()
    std=X.iloc[0+k:480+k].std()
    Mean.append(mean)
    Median.append(median)
    Max.append(maximum)
    Min.append(minimum)
    Std.append(std)

Min=pd.DataFrame(Min)
Min=Min.rename(columns={"avg_rss12":"Min1" ,  "var_rss12":"Min2","avg_rss13":"Min3" ,\
                     "var_rss13":"Min4", "avg_rss23":"Min5",  "var_rss23":"Min6"})
Min=Min.iloc[: , 1:]

Mean=pd.DataFrame(Mean)
Mean=Mean.rename(columns={"avg_rss12": "Mean1", "var_rss12": "Mean2","avg_rss13": "Mean3",\
                    "var_rss13": "Mean4","avg_rss23": "Mean5", "var_rss23": "Mean6"})
Mean=Mean.iloc[: , 1:]
Median=pd.DataFrame(Median)
Median=Median.rename(columns={"avg_rss12": "Median1", "var_rss12": "Median2","avg_rss13": "Median3",\
                    "var_rss13": "Median4","avg_rss23": "Median5", "var_rss23": "Median6"})
Median=Median.iloc[: , 1:]

Max=pd.DataFrame(Max)
Max=Max.rename(columns={"avg_rss12": "Max1", "var_rss12": "Max2","avg_rss13": "Max3",\
                    "var_rss13": "Max4","avg_rss23": "Max5", "var_rss23": "Max6"})
Max=Max.iloc[: , 1:]

Std=pd.DataFrame(Std)
Std=Std.rename(columns={"avg_rss12": "Std1", "var_rss12": "Std2","avg_rss13": "Std3",\
                    "var_rss13": "Std4","avg_rss23": "Std5", "var_rss23": "Std6"})
Std=Std.iloc[: , 1:]

```

```python
# 1st_Quartile
stQ=[]
B=[]

for j in range(6):
    for i, k in enumerate(neighbors):
        A=X.iloc[0+k:480+k,j+1]
        B.append(np.percentile(A, 25))
        stQ=pd.DataFrame(B)
stQ
A=np.array_split(stQ, 6)
A0=pd.DataFrame(A[0].values)
A1=pd.DataFrame(A[1].values)
First_Quart=pd.concat([pd.DataFrame(A[0].values),pd.DataFrame(A[1].values),\
          pd.DataFrame(A[2].values),pd.DataFrame(A[3].values),\
          pd.DataFrame(A[4].values),pd.DataFrame(A[5].values)], axis=1)
First_Quart.columns=['1st quart1','1st quart2','1st quart3','1st quart4','1st quart5','1st quart6']
 
First_Quart
```

```python
from scipy.stats import bootstrap
import numpy as np
Lo=[]
Hi=[]
CI_Lo=[]
CI_Hi=[]
STD=np.std(df)

for i in range(42):
    DD=df.iloc[:, i].to_list()
 
    data=(DD,)
    #print(data)
    res=bootstrap(data,np.std,random_state=24,confidence_level=0.9,method="percentile")
    Lo=res.confidence_interval[0]
    CI_Lo=np.append(CI_Lo,Lo)
    Hi=res.confidence_interval[1]
    CI_Hi=np.append(CI_Hi,Hi)
    #print(CI_Hi)
```
